```python
%matplotlib inline
import random
position=0
walk=[position]
steps=1000
for i in range(steps):
    step=1 if random.randint(0,1) else -1
    position+=step
    walk.append(position)
plt.plot(walk[:100])
#plt.plot(np.sin(walk[:50]))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
nstep=1000
draws=np.random.randint(0,2,size=nstep)
steps=np.where(draws>0,1,-1)
walk=steps.cumsum()
np.min(walk)
np.max(walk)
plt.plot(walk[:100])
(np.abs(walk)>=10).argmax()#argmax()返回最大值的第一个位置，在这里np.abs(walk)>=10会返回一个布尔数组（由True,False组成）所以最大值其实是True
                    #事实上，在这里使用argmax()，效率不高，因为它总是扫描整个数组。但是在这个布尔数组中，一旦扫描到True，就知道了最大值
```
